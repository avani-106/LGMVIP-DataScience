# Task-2 : Image to Pencil Sketch with Python


```python
import cv2
from PIL import Image
from IPython.display import display
#from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
```

# Loading the image 

```python
img = cv2.imread("C:\\Users\\HP\\Desktop\\photo-1544005313-94ddf0286df2.jpg")
img = cv2.resize(img, (600, 500))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Convert the image to gray scale

```python
gray_image = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
display(Image.fromarray(gray_image))

```

# Invert the image 

```python
inverted_gray_image = 255 - gray_image 
display(Image.fromarray(inverted_gray_image))

```

# Blur the image by gaussian function

```python
blurred_img = cv2.GaussianBlur(inverted_gray_image, (21,21), 0)
display(Image.fromarray(blurred_img))
```

# Invert the blurred image

```python
inverted_blurred_img = 255 - blurred_img
display(Image.fromarray(inverted_blurred_img))
```

# Create the pencil sketch image

```python
pencil_sketch_img = cv2.divide(gray_image, inverted_blurred_img, scale=256.0 )
display(Image.fromarray(pencil_sketch_img))
```
